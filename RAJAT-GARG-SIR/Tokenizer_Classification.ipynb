```python
import pandas as pd

data = [
    ("book a flight","booking"),
    ("cancel my ticket","cancel"),
    ("what is the weather","weather"),
    ("book a hotel","booking "),
    ("cancel reservation","cancel"),
    ("weather today","weather")
]

df = pd.DataFrame(data, columns=["text", "intent"])

df.head()
```

```python
import re
def custom_tokenizer(text):
  text = text.lower()
  text = re.sub(r"[^a-z0-9]"," ",text)
  tokens = text.split()
  return tokens

df["tokens"] = df["text"].apply(custom_tokenizer)
print(df)
```

```python
import numpy as vectorize
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(tokenizer=custom_tokenizer)
X = vectorizer.fit_transform(df["text"])
y = df["intent"]
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X,y)
```

```python
test_sentences = [
    "book a flight",
    "cancel my ticket",
    "weather tomorrow",
    "it's very hot",
    "raining in river",
    "cancel my class"
]

preds = model.predict(vectorizer.transform(test_sentences))
print(preds)
```

```python
# Mini BPE implementation

words = {
    "a n v a y":5,
    "a n v a y a":2,
}

def get_stats(words):
  pairs = {}
  for word, freq in words.items():
    chars = word.split()
    for i in range(len(chars) - 1):
      pair = (chars[i], chars[i+1])
      pairs[pair] = pairs.get(pair, 0) + freq
  return pairs


output = get_stats(words)
print(output)
```
